```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
all_data = []
names = ['delaunay80','globe','nefertiti','ngon10Q','orthogonal','random20','random40','small','sphere']
for i in names:
    s = 'json/' + i + '.json'
    with open(s) as json_file:
        all_data.append(json.load(json_file))
```

```python
all_data[0]
```

```python
def show(data,print_vertex = False):
    nodes_x = [dic['x'] for dic in data['nodes']]
    nodes_y = [dic['y'] for dic in data['nodes']]
    if print_vertex:
        print(nodes_x)
        print(nodes_y)
    plt.scatter(nodes_x,nodes_y)
    for i in data['edges']:
        plt.plot([nodes_x[i['source']],nodes_x[i['target']]],[nodes_y[i['source']],nodes_y[i['target']]])
    plt.grid()
    plt.show()
```

```python
for i in all_data:
    show(i)
```
