```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
all_data = []
#names = ['delaunay80','globe','nefertiti','ngon10Q','orthogonal','random20','random40','small','sphere']
names = ['example','orthogonal','random20','random40','small','sphere']
for i in names:
    s = 'json/' + i + '.json'
    with open(s) as json_file:
        all_data.append(json.load(json_file))
```

```python
def show(data,print_vertex = False):
    nodes_x = [dic['x'] for dic in data['nodes']]
    nodes_y = [dic['y'] for dic in data['nodes']]
    if print_vertex:
        print(nodes_x)
        print(nodes_y)
    plt.scatter(nodes_x,nodes_y)
    dist = np.array([])              # list of all distances
    for i in data['edges']:
        x_a = nodes_x[i['source']]
        x_b = nodes_x[i['target']]
        y_a = nodes_y[i['source']]
        y_b = nodes_y[i['target']]
        a = np.array([x_a,y_a])
        b = np.array([x_b,y_b])
        dist_ab = 0                    # compute the total distance between a and b
        if 'bends' in i:               # draw additional edges and update the distance
            x_c = i['bends'][0]['x']
            y_c = i['bends'][0]['y']
            plt.plot([x_a,x_c],[y_a,y_c])
            c = np.array([x_c,y_c])
            dist_ab += np.linalg.norm(a-c)
            n = len(i['bends'])
            for j in range(0,n-1):
                x_p0 = i['bends'][j]['x']
                y_p0 = i['bends'][j]['y']
                x_p1 = i['bends'][j+1]['x']
                y_p1 = i['bends'][j+1]['y']
                plt.plot([x_p0,x_p1],[y_p0,y_p1])
                dist_ab += np.linalg.norm(np.array([x_p0,y_p0])-np.array([x_p1,y_p1]))
            x_d = i['bends'][n-1]['x']
            y_d = i['bends'][n-1]['y']
            plt.plot([x_d,x_b],[y_d,y_b])
            d = np.array([x_d,y_d])
            dist_ab += np.linalg.norm(b-d)
        else:
            plt.plot([x_a,x_b],[y_a,y_b])
            dist_ab = np.linalg.norm(a-b)
        dist = np.append(dist,dist_ab)
    
    # find max and min
    dist_min = np.amin(dist)
    dist_max = np.amax(dist)
    plt.grid()
    plt.xlim(-1, data['width']+1)
    plt.ylim(-1, data['height']+1)
    plt.show()
    if dist_min != 0:
        print('Edge length ratio: ', dist_max/dist_min)
```

```python
for i in all_data:
    show(i)
```
